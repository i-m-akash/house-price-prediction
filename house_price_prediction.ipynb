```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/drive/MyDrive/ML/kc_house_data.csv')
print(df)
```

```python
df=df.drop(columns='date',axis=1)
df
```

```python
df.isnull().sum()
```

```python
X=df.drop(columns='price',axis=1)
Y=df['price']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
#Split the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state=2)
```

```python

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

# get numeric data
num_v = df.select_dtypes(exclude=['object'])

# update the cols with their normalized values
df[num_v.columns] = sc.fit_transform(num_v)

# convert string variable to One Hot Encoding
#Handling categorical features
df1 = pd.get_dummies(df)

df1.head(10)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=model.score(X_train,Y_train)
training_data_accuracy
```

```python
X_test_prediction=model.predict(X_test)
testing_data_accuracy=model.score(X_test,Y_test)
testing_data_accuracy
```

```python
from sklearn.metrics import mean_squared_error
rmse_test=mean_squared_error(X_test_prediction,Y_test)**0.5
print(rmse_test)
```
